```python
merged['HD01_VD01'] = merged['HD01_VD01'].astype(int)

merged.HD01_VD01
```

```python
listpoly = []

for i in range (11028):
    if merged.HD01_VD01[i] == 0:
        listpoly.append(0)
        
    else:
        
        polygon = merged.geometry[i]
    
        count = 0
        for j in range (52055):
            point = Point (collisions.X[j], collisions.Y[j])
        
            if polygon.contains(point):
                count += 1
    
        listpoly.append(count)
    
listpoly
```

```python
merged['contained'] = listpoly
```

```python
count = 0
for i in range(11028):
    count += merged.contained[i]
    
count
```

```python
copy2 = df
indexlist = []

#for i in range (292):
#    if (copy2.collision[i] == 0) or (copy2.collision[i] <= 30 and copy2.population[i] >= 1000):
#        indexlist.append(i)

for i in range (292):
    if (copy2.collision[i] == 0):
        indexlist.append(i)

        
indexlist
```

```python
for i in range (292):
    print (copy2.population[i], \t,copy2.collision[i])
```

```python
copy3 = copy2.drop(indexlist)

copy3
```

```python
plt.scatter (copy3.collision, copy3.population)
plt.xlabel ('Number of collision')
plt.ylabel ('Population')
plt.show()
```

```python
from sklearn.cluster import KMeans
est = KMeans(3)  # 3 clusters
est.fit(copy3)
y_kmeans = est.predict(copy3)
plt.scatter(copy3.collision, copy3.population, c=y_kmeans, cmap='rainbow');
plt.xlabel ('Number of collision')
plt.ylabel ('Population')
```
